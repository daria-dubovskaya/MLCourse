1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = sp.symbols('x')
f = 10 * x - 5
derivative_f = sp.diff(f, x)
derivative_f
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 400)
f = 10 * x - 5
df = np.full_like(x, 10)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Синя лінія є графіком функції та показує як залежність оцінки від кількості годин проведених за навчанням.Функція є лінійною з нахилом 10 - це означає, що з кожною витраченою годиною на навчання, результат поліпшується на 10 балів.

Зелена лінія показує похідну функції, вона є постійною і дорівнює 10.Вона показує, що швидкість зміни результату навчання є постійною і дорівнює 10 балів на годину.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f = sp.sin(x)
derivative_f_sin = sp.diff(f, x)

x = sp.symbols('x')
f = sp.ln(x)
derivative_f_ln = sp.diff(f, x)

derivative_f_sin, derivative_f_ln
```

```python
xs = np.linspace(0, 10, 400)
f_sin = np.sin(xs)
df_sin = np.cos(xs)
plt.plot(xs, f_sin, label='f(x) = sin(x)')
plt.plot(xs, df_sin, label='f(x) = cos(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
xs = np.linspace(0.1, 10, 400)
f_ln = np.log(xs)
df_ln = 1 / xs
plt.plot(xs, f_ln, label='f(x) = ln(x)')
plt.plot(xs, df_ln, label='f(x) = 1 / x')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
